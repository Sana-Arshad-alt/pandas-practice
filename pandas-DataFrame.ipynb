```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame([
   ['Sana', 22],
    ['Sadia',24],
    ['Sania',27],
    ['Salma',30]
    ] , columns=['Name', 'Age'])
plt.plot(df['Name'],df['Age'] , color = 'blue')
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Ages of Students")
plt.grid(True)
plt.show()
```
